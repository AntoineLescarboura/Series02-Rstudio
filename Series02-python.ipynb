## The goal of this notebook is to plot sinus signal as somethin useful 
 Math fotmula of the sinusoid = y(t) = A * sin (2 pi ft + φ)

 A = amplitude 
 f = frequency of the signal (in Hz)
 t = time (seconds)
 φ = initial phase 

```python
import numpy as np 
import matplotlib.pyplot as plt

#Create a numpy array called t that contains 1000 values from 0 to 1 (t is for time)
t = np.array([i/1000 for i in range(1001)])

#Create a numpy array called s that contains the values of a sine function of t (s is for signal)
s = np.sin(2 * np.pi * 2 * t) # 1 Hz sine wave

#Plot the signal s as a function of time t
plt.plot(t, s)
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.title('Sinusoidal Function')
plt.grid()
plt.savefig('sine_wave.png')
plt.show()

```

# Creation of the sinusoid 

```python
import numpy as np
import matplotlib.pyplot as plt

def sinusoid (amplitude : float, frequency: float, duration: float, num_points: int=1000, phase: float = 0.0):
#     Create a sinusoidal signal

#     Parameters
#     ---------
# float = real number with decimal 

#     amplitud :(maximal amplitude of the sinus)
#     frequency : (frequency in Hz)
#     duration : total time of the signal in seconds
#     num_points : number of the signal points (default is 1000)
#     phase : inital phase in radians 

#     Returns 
#     -------
#     np.ndarray = sinusoid signal evaluated on a linear vector 

    t=np.linspace (0, duration, num_points) # time vector 
    x= amplitude * np.sin(2*np.pi*frequency*t+phase)
    return x,t 
```

<br>

# Creating two signals with different frequencies

```python
# Parameters definition
amplitude = 1.0
duration = 1.0  # seconds
num_points = 1000

# Different frequencies
freq1 = 1  # Hz
freq2 = 3  # Hz

# Signal generation
sine_wave_1, t = sinusoid(amplitude, freq1, duration, num_points)
sine_wave_2, _ = sinusoid(amplitude, freq2, duration, num_points)  # time vector already calculated
```

<br>

# Plot the two signals 

```python

# Plot the two signals
plt.plot(t, sine_wave_1, label=f"Sine 1")
plt.plot(t, sine_wave_2, label=f"Sine 2")

# Labels and titles
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.title("Comparison of Two Sinusoids with Different Frequencies")
plt.grid(True)
plt.legend()

# Saving the figure
plt.savefig("two_sinusoids.pdf")
plt.show()
```

# 3.1 Even or odd

### Even number : is an integer that is divisible by 2 without a remainder

```python
import numpy as np 

# create a series of number from 0 to 10 and print if it is even or odd
numbers = np.arange(0, 11) # create a numpy array from 0 to 10

#create a loop
for number in numbers : # loop through each number in the array
    if number % 2 == 0: # check if the number is even by dividing by 2 and checking the remainder
        print(f"{number} is even")
    else:
        print(f"{number} is odd")
```

# Multiple or not 

### A **multiple** of a number \(m\) is any integer that can be expressed as the product of \(m\) and another integer \(k\).

Mathematically, a number \(n\) is a multiple of \(m\) if:

\[
n = m \cdot k \quad \text{where } k \text{ is an integer.}
\]

```python
import numpy as np

# create a series of numbers from 0 to 20
numbers = np.arange(0, 21) # create a numpy array from 0 to 20

#create a loop to check if multiples of 3 or 5 
def check_multiple(number: int, m: int):
    if number % m == 0:
        print(f"{number} is a multiple of {m}")
    else:
        print(f"{number} is NOT a multiple of {m}")

numbers = range(0, 21)  # 0 à 20
for number in numbers:
    check_multiple(number, 3)  # vérifie multiple de 3
    check_multiple(number, 5)  # vérifie multiple de 5
    print()  # ligne vide pour séparer les nombres
```

# Prime or not 

### A **prime number** is an integer greater than 1 that has **exactly two positive divisors**: 1 and itself.  

Mathematically, a number \(p\) is prime if:

\[
\text{Divisors of } p = \{1, p\}
\]

```python
def is_prime(n) : 
    if n < 2 :
        return False
    for i in range(2, n) :
        if n % i == 0 :
            return False
    return True
numbers = range(0, 200)  # 0 à 200   
primes_200 = [n for n in numbers if is_prime(n)]
print(primes_200)  # Affiche les nombres premiers entre 0 et 20
```
